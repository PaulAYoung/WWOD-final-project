# California Education Data Quickstart

In this notebook we will load a list of schools and do some simple filtering of it. 

Other notebooks will require you to download additional data sets. 

```python
import os
import pandas as pd
```

```python
#get data dir

data_path = os.path.join(os.path.pardir, "data")
assert os.path.exists(data_path)
```

```python
#load school list data and show first few rows
school_list = pd.read_csv(os.path.join(data_path, "pubschls.txt"), sep="\t")
school_list.head()
```

```python
#show columns
school_list.columns
```

```python
# get list schools in Berkeley unified
berkley_schools = school_list[school_list.District == "Berkeley Unified"]
berkley_schools.head()
```

```python
#show the different types of schools that are open in each district
# the column meanings can be found at http://www.cde.ca.gov/ds/si/ds/fspubschls.asp
# under the EdOpsCode variable
school_list.pivot_table(values='School', rows=['County'], cols=['EdOpsCode'], aggfunc='count')
```
