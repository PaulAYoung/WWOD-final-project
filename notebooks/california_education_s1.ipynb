### Section 3: Subject Data

Objective: What kind of subjects are being offered in schools around California? What are some of the trends around gender distribution for the courses being offered? What are some schools offering in terms of courses and how does that differ between different schools?

I used three distinct data files and compiled them into a single data frame. This data frame stores all the information relating the school, its charter and non-charter characteristic, subjects offered, number of students enrolled and some more granular details about the students involved. 

I approached the problem by aggregating the majority of the data together into one data frame and then slowly slicing the main data frame, dissecting the frame into separate chunks for analysis. You can see in the link to the notebook for more detail on this process.

After some data exploration into the names of the school, the population of students enrolled in the school, I found it interesting to show some basis findings on the gender distribution and summary of the subjects offered by school.


```python
#read public schools table
public_school = pd.read_csv(os.path.join(data_path, "pubschls.txt"), sep="\t")

#read Assignment 10 csv data
assign10 = pd.read_csv(os.path.join(data_path, 'assign10.txt'), sep="\t")

#public schools to join with staff assignment table by CDS_CODE
merge_public_assign10 = public_school.merge(assign10, left_on="CDSCode", right_on="CDS_CODE")

#read Course Subject breakdown
assignmentcolumns = [
    "ASGN_CODE",
    "ASGN_NAME",
    "CTYPE",
    "SUBJ_CODE",
    "SUBJECT",
    "TOPIC_CODE",
    "TOPIC_HEAD",
    "START_DT",
    "LAST_DT",
    "UC_CSU_REQ",
    "Comments"
]
subjectdata = pd.read_csv(os.path.join(data_path, "asgncode.txt"), sep="\t", header=None, names=assignmentcolumns)

#staff assignment table to course by the asn_code
#merge assign10 to subjectdata
merge_all = merge_public_assign10.merge(subjectdata, left_on="ASN_CODE", right_on="ASGN_CODE")
merge_all = merge_all[['CDSCode','County','District','School','SCHOOOL', 'Charter', 'FundingType',
           'ASGN_NAME', 'ASN_CODE', 'M_ENROLL', 'F_ENROLL', 'SUBJECT', 
           'TOPIC_HEAD', 'START_DT', 'LAST_DT']]
merge_all_for_chart = merge_all[['County','District','SCHOOOL','ASGN_NAME','SUBJECT', 
           'TOPIC_HEAD']].sort(['County','District','SCHOOOL','SUBJECT','TOPIC_HEAD','ASGN_NAME'],ascending=True)

#Quick Summary of the Overall Trends
#summary of data
merge_all_for_chart.describe()
```

```python
#all subjects across the entire California
merge_all_subjects = merge_all[['SUBJECT','F_ENROLL', 'M_ENROLL']]
merge_all_subjects_groupby = merge_all_subjects.groupby('SUBJECT')

#merge_all_subjects_groupby.get_group('english').F_ENROLL.sum()
merge_all_schools = merge_all[['County','SUBJECT','F_ENROLL', 'M_ENROLL']]
merge_all_schools_groupby = merge_all_schools.groupby(['County', 'SUBJECT'])

#create dropdown widget
subject_widget = (merge_all_subjects['SUBJECT'].unique()).tolist()
#enter values into the widget
subject_widget_filter = widgets.DropdownWidget(values=subject_widget)

#create function that executes whenever the value of the widget changes
def widget_change_for_subject(Subject):
    merge_all_subjects_groupby.get_group(Subject)[["F_ENROLL", "M_ENROLL"]].sum().plot(kind='bar', figsize=(8,5), title=Subject, color='c')
    
interact(
    widget_change_for_subject, 
    Subject = subject_widget,
) 
```
