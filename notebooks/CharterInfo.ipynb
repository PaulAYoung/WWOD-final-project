# California Charter Schools and Dropout Rates

###Overview

This notebook examines the relationship between the adoption of charter schools in California and school performance. Charter schools have been steadily increasing in California over the past 20 years but their affect on an education system is still unclear. There have been rave reviews of certain charter schools as boosting performance and being a panacea to education concerns. Critics say that the record of charters is not so clear and they often attract students that are more motivated or have more involved parents and leave the publics schools to deal with the more difficult children. 

I wanted to investigate this and see if there was any evidence for or against them. I chose to look at at dropout rates over the district as a whole to test the overall impact of charter schools. 

####Findings
I found that there was a slight correlation between districts that adopted charter schools at a quicker rate and an increase in dropout rate. However, it is difficult to show directionality in this relationship as districts may be pushed to adopt charter schools as a way of combating dropouts. 

Given the data it appears that charter schools are not an effective way to reduce the dropout rate in a school district.

#Correlation of Charter use and Dropoutrate
The following code tries to find a statistical link between charter adoption and changes in dropout rate at the district level. 

```python
#Calculate the trend (slope of the line of best fit) for charter use and dropout rates in each district. 

dist_drop.fillna(0, inplace=True)
change = []
for c in dist_drop.index.levels[0].unique():
    for d in dist_drop.ix[c].index.get_level_values(0).unique():
        #print c + " - " + d
        x = dist_drop.ix[c, d].reset_index()["year"]
        y = dist_drop.ix[c, d]["All"]
        if len(y) < 10:
            continue
        model = np.polyfit(x=x, y=y, deg=1)
        row = {
            "County": c,
            "District": d,
            "dropout_beta": model[0]
        }
        x = dist_enr.ix[c, d].reset_index()["year"]
        y = dist_enr.ix[c, d]["All"]
        model = np.polyfit(x=x, y=y, deg=1)
        row["charter_beta"] = model[0]
        change.append(row)
        
        
change_df = pd.DataFrame(change)

del change
```

```python
#clean the data by removing na and infinity values
change_df.dropna(inplace=True)
change_df = change_df[~change_df.dropout_beta.isin([np.inf, -np.inf])]
change_df = change_df[~change_df.charter_beta.isin([np.inf, -np.inf])]
```

###Model
The following creates a model to check for a correlation between charter adoption and changes in dropout rate. 

```python
#create model to show relationship
model = pd.ols(x={"charter": change_df.charter_beta}, y=change_df.dropout_beta)
model
```

This model shows a correlation between a quicker pace of charter adoption and an increase in the dropout rate. Lets see what the data looks like. 

```python
plt.scatter(change_df.charter_beta, change_df.dropout_beta)
```

looks like there are a few outliers. Let's make a new model but try to only include districts with atleast 5000 high school students and a dropout change rate of less than .05. 

```python
d_size = 5000 # minimum number of high school students
drop_rate = .05 # the highest change in dropout rate allowed to filter outliers. 
big_dists = enrollment[enrollment.All > d_size].index.get_level_values(1).unique()
f_change = change_df[change_df.District.isin(big_dists)]
f_change = f_change[f_change.dropout_beta < drop_rate]
```

```python
model = pd.ols(x={"charter": f_change.charter_beta}, y=f_change.dropout_beta)
model
```

```python
plt.scatter(f_change.charter_beta, f_change.dropout_beta)
```

####Analysis
The model still shows faster adoption of charter schools correlating with an increase in the dropout rate, however a visual check on the data shows that there are a few outliers that may be dragging the model. I will leave it to the user to play with the filters to try and eliminate outliers.

#Conclusion
Charter schools certainly do not appear to be improving charter rates and there is even a correlation between quicker adoption of charter schools and an increase in the dropout rate. This correlation is likely because charter schools are adopted in reaction to higher drop out rates, not because charters themselves cause dropouts. However, given this data it appears that charter schools are not an effective means to address high school dropouts.
