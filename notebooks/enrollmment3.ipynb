```python
%matplotlib inline

import os
import pandas as pd
```

```python
from IPython.html.widgets import interact, interactive, fixed
from IPython.html import widgets
from IPython.display import clear_output, display, HTML
import matplotlib.pyplot as plt
```

```python
#get data dir

data_path = os.path.join(os.path.pardir, "data")
compiled_data = os.path.join(os.path.pardir, "compileddata")
final_data = os.path.join(os.path.pardir, "finaldata")
assert os.path.exists(data_path)
assert os.path.exists(compiled_data)
assert os.path.exists(final_data)
```

```python
enr = pd.read_pickle(os.path.join(final_data, "p_charter_county_dist.pkl"))
```

```python
enr.ix["Alameda"].index.get_level_values(0).unique()
```

```python
county_select = widgets.DropdownWidget(values=list(enr.index.get_level_values(0).unique()))
district_select = widgets.DropdownWidget()

def select_districts(trait, value):
    district_select.values={i: i for i in list(enr.ix[value].index.get_level_values(0).unique())}

county_select.on_trait_change(select_districts, 'value')
```

```python
ethnic_groups = list(enr.columns.get_level_values(1).unique())
areas = list(enr.index.get_level_values(0).unique()),
ethnic_checkboxes = []
for g in ethnic_groups:
    ethnic_checkboxes.append(widgets.CheckboxWidget(description=g, value=True))
```

```python
def plot_district(county, district, **kwargs):
    #set group colors
    colors = [
    "red",
    "orange",
    "yellow",
    "green",
    "blue",
    "purple",
    
]
    
    groups = []
    fig = plt.figure()
    ax = fig.add_axes([0,0,2,1])
    ax.set_ylim(0,1)
    ax.set_xlim(1990, 2015)
    
    for k, v in kwargs.items():
        if v:
            kwargs = {}
            if k == "All":
                kwargs["color"] = "black"
                kwargs["figsize"]=(6,6)
            enr.ix[county, district][k].plot(**kwargs)
    plt.legend(loc='best')
```

```python
ethnicity_checkboxes = {
    g: widgets.CheckboxWidget(description=g, value=True) for g in ethnic_groups
}

interactive(
    plot_district,
    county = county_select,
    district = district_select,
    **ethnicity_checkboxes
)
```
